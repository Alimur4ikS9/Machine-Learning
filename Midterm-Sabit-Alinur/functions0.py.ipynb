```python
#Task 1
x = [[1, 0, 1], [2, 0, 2], [3, 0, 3], [4, 4, 4]] 
result = 1
for i in range(len(x)): 
    for j in range(len(x[i])):
        if(i == j and x[i][j] != 0):
            result *= x[i][j]
result
```

```python
#Task 4
import matplotlib.pyplot as pt
import matplotlib.image as im
coefs = [0.299, 0.587, 0.114]
def convert_image(img, coefs):
    for x in range(len(img)):
        for y in range(len(img[x])):
            img[x, y] = (img[x][y][0] * coefs[0] + 
                         img[x][y][1] * coefs[1] + 
                         img[x][y][2] * coefs[2])
    return img
    pass
img = im.imread('zhabka.png')
gray = convert_image(img, coefs)
pt.imshow(gray, cmap = pt.get_cmap('gray'))
pt.show()
```
