### Задача 2: кросс-валидация для исходных данных
__(0.5 балла)__

Загрузите файл scores.csv и протестируйте, как линейная регрессия предсказывает ответ сейчас (с помощью кросс-валидации).

*Hint: воспользуйтесь sklearn.linear_model и sklearn.model_selection.*

```python
scores = pd.read_csv("scores.csv")
```

```python
from sklearn.model_selection import train_test_split #split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(data.iloc[1:,:], scores, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
linReg = LinearRegression() #instantiating 
linReg.fit(X_train,y_train) #fit linear model
linReg.predict(X_test) #predict using the linear model
"""Техника валидации модели для проверки того, 
    насколько успешно применяемый в модели статистический анализ 
    способен работать на независимом наборе данных.
    Нужно оценить, насколько предсказывающая модель способна работать на практике"""
```

```python
from sklearn.metrics import mean_squared_error
(mean_squared_error(y_test, linReg.predict(X_test)))**(1/2)
```
