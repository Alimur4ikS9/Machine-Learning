```python
#Task 3
x = np.array([6, 2, 0, 19, 0, 0, 10, 7, 0])
indices = np.where(x == 0) 
indices = np.add(indices, np.array([1]))
indices = indices[indices != len(x)] #удаляет индекс больше длины (Например, последний элемент 0, поэтому indices + 1 не сможем сделать)
np.amax(np.take(x, indices)) #макс элемент в массиве между индексами(полученными выше)
```

```python
#Task 4
coefs = np.array([0.299, 0.587, 0.114])
def convert_image(img, coefs):
    return np.dot(img[:,:,:3], coefs)

img = im.imread('ironMan.png')
gray = convert_image(img, coefs)
casual = pt.figure('Vectorized') 
pt.imshow(gray, cmap = 'Greys_r')
#iw('gray_zhabka.png', convert_image(img, coefs))
```
